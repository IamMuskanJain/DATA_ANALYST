### Task 1: Preparing for Precision
In our quest to enhance maternal and fetal healthcare, our journey has led us to a pivotal juncture: the predictive analysis of fetal health. This endeavor is fueled by a deep-rooted commitment to ensure the well-being of expectant mothers and their precious unborn children. We recognize that data-driven insights hold the key to achieving this noble objective.

At this stage, our primary task is to divide our dataset into distinct training and testing sets. This strategic division is essential to develop and evaluate predictive models that can accurately determine fetal health. The gravity of our mission underscores every decision we make, and the outcome of our analysis carries the potential to offer invaluable reassurance and medical insights to expectant mothers, empowering them to make informed decisions about their pregnancies. In this journey, we find purpose and the opportunity to make a meaningful impact on the lives of those we aim to serve.

```python
# ...WRITE YOUR CODE FOR TASK 1 ...
from sklearn.model_selection import train_test_split
# Inspect data
Xx = df.drop(["fetal_health"], axis=1)
Yy = df["fetal_health"]
x_train, x_test, y_train, y_test = train_test_split(Xx,Yy,test_size=0.3, random_state=42)
x_train, x_test, y_train, y_test
```

### Task 2: Precision with Logistic Regression
In our mission to predict fetal health with precision, we turn to the world of machine learning, where Logistic Regression emerges as our chosen ally. With "liblinear" as the solver, we train our Logistic Regression model, aptly named model1b, on the training data. This phase of learning is critical, as it allows the model to discern intricate patterns within the dataset, connecting the dots between various attributes and fetal health classifications.

The true litmus test arrives as we apply our trained model to the test set, x_test, to generate predictions. It is here that we measure the model's prowess through the prism of accuracy. The accuracy score is more than just a metric; it is the embodiment of our dedication to precision and the cornerstone of our efforts to provide accurate fetal health predictions. With every accurate prediction, we take a significant stride toward a future where expectant mothers can rely on data-driven insights to make informed decisions about the well-being of their unborn children.

```python
# ...WRITE YOUR CODE FOR TASK 2 ...
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model1b = LogisticRegression(solver='liblinear').fit(x_train,y_train)
y_predict1b = model1b.predict(x_test)
accuracy_lr = accuracy_score(y_test, y_predict1b)
# Inspect data
accuracy_lr
```

### Task 3: Prediction with RandomForestClassifier
In our ongoing quest to enhance the accuracy of fetal health predictions, we introduce the powerful RandomForestClassifier to our toolkit. This model, celebrated for its ensemble of decision trees and predictive prowess, serves as a crucial addition to our mission.

As we name our model, model2b, and set it on its path of learning, we understand that the essence of its capabilities lies in its ability to decipher complex patterns hidden within the dataset. It takes on the role of a diligent student, meticulously studying the attributes and the relationships within. With each decision tree contributing its unique perspective, our model becomes a formidable entity, ready to unlock the intricacies of fetal health.

Turning our attention to the test dataset, x_test, we aim to assess the model's predictive abilities. The predictions, denoted as y_pred2b, undergo the scrutiny of the accuracy score. Beyond being a numerical metric, this score symbolizes our unwavering commitment to harness advanced machine learning techniques in our pursuit of precise fetal health predictions. With every high-accuracy prediction, we move one step closer to a future where expectant mothers can rely on data-driven insights to make well-informed decisions about the well-being of their unborn children.

```python
# ...WRITE YOUR CODE FOR TASK 3 ...
from sklearn.ensemble import RandomForestClassifier
model2b = RandomForestClassifier(criterion = 'gini',n_estimators=100,max_depth=4,random_state=33) #criterion can be also : entropy
model2b.fit(x_train, y_train)
y_pred2b = model2b.predict(x_test)
accuracy_rf = accuracy_score(y_test, y_pred2b)
accuracy_rf
# Inspect data
```

### Task 4: Prediction with DecisionTreeClassifier
In our relentless pursuit of precise fetal health predictions, we introduce the DecisionTreeClassifier into our ensemble of models. Renowned for its ability to make decisions based on intricate data patterns, this classifier offers a different perspective in our quest.

Model3b, our DecisionTreeClassifier, is carefully configured to strike a balance between simplicity and predictive power. It sets out on its learning journey, much like a perceptive student, aiming to decipher the key attributes and patterns that influence fetal health outcomes.

As we turn our focus to the test dataset, x_test, we set out to evaluate the DecisionTreeClassifier's predictive capabilities. The predictions, denoted as y_pred3b, undergo the litmus test of the accuracy score. Beyond being a numerical measure, this score signifies our commitment to exploring diverse machine learning approaches in our mission for precise fetal health predictions. With each accurate prediction, we take another step toward a future where expectant mothers can make well-informed decisions about their unborn children's well-being, guided by the insights derived from advanced data analytics and machine learning.

```python
# ...WRITE YOUR CODE FOR TASK 4 ...
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model3b = DecisionTreeClassifier(criterion='gini', max_depth=2, random_state=33)
model3b.fit(x_train, y_train)
y_pred3b = model3b.predict(x_test)
accuracy_dt = accuracy_score(y_test, y_pred3b)
accuracy_dt
```
