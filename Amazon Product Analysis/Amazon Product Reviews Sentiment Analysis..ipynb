**Task 4: Exploring Amazon Product Reviews.**

In this pivotal chapter of our sentiment analysis journey, I've harnessed NLTK's SentimentIntensityAnalyzer to unveil the intricate emotions within Amazon product reviews. Just as a skilled director uncovers the depths of their characters, I've introduced "Positive," "Negative," and "Neutral" columns, providing a multidimensional perspective to the narrative. These columns serve as the main characters in our data story, capturing the sentiment nuances within each review.

```python

import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentiments = SentimentIntensityAnalyzer()
# --- WRITE YOUR CODE FOR TASK 4 ---
df["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in df["Text"]]
df["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in df["Text"]]
df["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df["Text"]]

#--- Inspect data ---
df
```

**Task 5: Unmasking Amazon's Dominant Emotion.**

In this project's defining moment, my code performs an emotional symphony, summing up the sentiments found within Amazon product reviews. Like a masterful film's climax, I've skillfully identified the dominant sentiment category, be it positive, negative, or neutral.

The variable 'sent' stands as the revelation, embodying the majority sentiment among these reviews, adding an exhilarating twist to our data journey. It's akin to uncovering the protagonist of a thrilling tale.

With this critical insight in hand, my narrative takes a bold step towards understanding the overarching sentiments within Amazon's vast sea of reviews, much like a movie unveiling the grand theme at its core. This is the turning point, and the ensuing scenes promise to be both enlightening and captivating.

```python
# --- WRITE YOUR CODE FOR TASK 5 ---
x = sum(df["Positive"])
y = sum(df["Negative"])
z = sum(df["Neutral"])
if (x>y) and (x>z):
    sentiment = "Positive"
elif (y>x) and (y>z):
    sentiment = "Negative"
else:
    sentiment = "Neutral"
sentiment
```
