Which region have the highest total sales revenue by region?

```python
HighestTotalRevenue = data.groupby(data['Region'])['Total Revenue'].sum()
HighestTotalRevenue
```

```python
HighestTotalRevenue.idxmax()
```

```python
Group_Data = data.groupby(data['Region'])['Total Revenue'].sum()
sns.set_style('darkgrid')
plt.figure(figsize=(15,5))
sns.lineplot(data = Group_Data, linestyle= '--', linewidth= 4, marker= 'o', markersize= 10, markerfacecolor= 'red', markeredgecolor= 'black')
plt.xticks(rotation= 90)
plt.title('Highest Total Sales Revenue by Region', fontsize = 20, color= 'blue')
plt.xlabel('Region', fontsize = 15, color= 'Red')
plt.ylabel('Total Revenue', fontsize = 15, color= 'Red')
plt.show()
```

Average unit cost and unit price for each item type

```python
AvgUnitPrice = data.groupby(data['Item Type'])['Unit Price'].mean()
AvgUnitCost = data.groupby(data['Item Type'])['Unit Cost'].mean()
UnitCostPrice = pd.DataFrame({'Average Unit Price': AvgUnitPrice, 'Average Unit Cost': AvgUnitCost})
UnitCostPrice
```

```python
UnitCostPrice.plot(kind='bar', stacked=False, figsize=(10, 5))
plt.title('Average unit cost and unit price for each Item')
plt.xlabel('Item Type')
plt.ylabel('Value')
plt.xticks(rotation=90)
# plt.legend(title='')
plt.show()
```

**Which country has the highest total profit?-- bar graph**

```python
total_profit = data.groupby(data['Country'])['Total Profit'].sum()

plt.figure(figsize=(25, 6))
plt.bar(x= total_profit.index, height= total_profit.values, color='skyblue')

plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.title('Total Profit by Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**How does the sales channel affect the order priority distribution?-- bar graph**

**What is the average order processiong time(duration between order and ship date) for each sales channel? --- bar plot**


**Which item types have the highest and lowest total sales? -- scatter plot**

```python
total_sales = data.groupby('Item Type')['Units Sold'].sum().reset_index()
total_sales
```

```python
highest_sales_item = total_sales.loc[total_sales['Units Sold'].idxmax()]
highest_sales_item
```

```python
lowest_sales_item = total_sales.loc[total_sales['Units Sold'].idxmin()]
lowest_sales_item
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(total_sales['Item Type'], total_sales['Units Sold'], color='blue')
plt.scatter(highest_sales_item['Item Type'], highest_sales_item['Units Sold'], color='green', label='Highest Sales')
plt.scatter(lowest_sales_item['Item Type'], lowest_sales_item['Units Sold'], color='red', label='Lowest Sales')

plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

**Are there any seasonal trends or pattern in the sales data**

```python
month_names = {1:'jan',2:'feb',3:'mar',4:'apr',5:'may',6:'jun',7:'jul',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'}
monthsales = data.groupby(data['Order Date'].dt.month)['Total Revenue'].sum()
monthsales
```

```python
monthsales.index = monthsales.index.map(month_names)
monthsales
```

```python
monthsales.plot(kind='bar', stacked=False, figsize=(10, 5))
plt.title('Seasonal trends')
plt.xlabel('Months')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
# plt.legend(title='')
plt.show()
```

```python
sns.barplot(x= monthsales.index, y= monthsales, hue= monthsales.index)
plt.title('Seasonal trends')
plt.xlabel('Months')
plt.ylabel('Total Revenue')
plt.show()
```

**Are there any outliers in the total cost distribution**

```python
q1 = data['Total Cost'].quantile(0.25)
q3 = data['Total Cost'].quantile(0.75)
iqr = q3-q1
lowerFence = q1-1.5*iqr
upperFence = q3+1.5*iqr
outliers = data[(data['Total Cost'] < lowerFence) | (data['Total Cost'] > upperFence)].reset_index(drop = True)
outliers
```

```python
plt.boxplot(data['Total Cost'])
plt.title('Box plot of total cost')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.boxplot(y='Total Cost', data=data, color='skyblue', width=0.5)
plt.title('Quantile Plot for Total Cost')
plt.ylabel('Total Cost')
plt.tight_layout()
plt.show()
```
